### Outlier(s) Detection:

### Removal of min_nights outliers with percentile and storing in the same dataframe df

```python
min_nig= df.minimum_nights.quantile(0.999)
min_nig
```

```python
df= df[df.minimum_nights<=min_nig]
sns.boxplot(data=df, y='minimum_nights')
```

```python
#Minimum nights outliers are removed as it make no senese to have 'Minimum nights >365 days'
#The individual would rather pick a leased property/rent property rather than booking through airbnb
```

# Bivariate Analysis
### Bivariate analysis is the analysis of the relationship between two variables. It helps to understand how one variable affects or is related to another

```python
viz = sns.barplot(data=df, x='neighbourhood_group', y='price per bed')
plt.title("Mean price per bed in the neighbourhood")
plt.xlabel("Neighbourhood Group   fig(c)")
plt.ylabel("Mean Price/Bed")
for lables in viz.containers:
    viz.bar_label(lables, fontsize=7, padding=3)
```

```python
#Manhattan being the most expensive area have the highsest mean listing price per bed of $140
#Followed by brooklyn at $100 and Queens standing at $77.
```

```python
df.columns
```

```python
viz = sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title("Mean price/bed & Dependency on Neighbourhood")
plt.xlabel("Neighbourhood Group   fig(d)")
plt.ylabel("Avg Listing price")
for labels in viz.containers:
    viz.bar_label(labels, fontsize=7, padding=3)
plt.show()
```

```python
#Deepdiving in to the room types of each area where mean price/listing is analysed
#Hotel listings are costliest in Manhatten whereas shared rooms and private rooms listings are low in cost comparitively
#Private room listings are chepest in Bronx followed by queens
#Listings in the Apartments/homes are chepest in Staten Island in the Newyork city
```

```python
#No of reviews and price relationship
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='price', y='number_of_reviews')
plt.title("Relationship of Price with Reviews")
plt.xlabel("price   fig(e)")
plt.show()
```

```python
#The scatter chart "Relationship of Price with Reviews" shows an inverse relationship between price and the number of reviews—higher-priced listings 
#tend to receive fewer reviews. This suggests that expensive properties attract a smaller, more selective group of guests. 
#In contrast, more affordable listings (priced between $1 and $750) have the highest number of reviews, indicating higher booking frequency 
#and broader appeal. This trend implies that budget-friendly options cater to a larger audience, while luxury listings are chosen 
#by a niche market with lower turnover.
```

```python
#No of reviews and price relationship
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='price', y='number_of_reviews', hue='room_type')
plt.title("Price with room_type and Review Dependency")
plt.xlabel("price   fig(f)")
plt.show()
#WE see reviews decreases as prices across rooms increases
```

### Heat Map - Correlation of one attribute with others

```python
df.columns
```

```python
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'availability_365', 'beds']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True )
plt.title("Correlation Heatmap")
plt.show()
```

```python
#PRICE and BEDS have a stong positive corr; where price increases as coun of beds in the listing increases
#A slight -ve tendency might exist btwn PRICE-Min_nights (as the price increases, the required minimum nights might decrease very slightly), 
#but it is so weak that it likely won’t have much practical significance in the data.
#Same goes with PRICE-Availability_365 but a slightly +ve relationship.(As price increases the availabilty of the listing also increases); as higher pries usually do not attract customers
```

# Conclusion/Insights:

- Price distribution across NYC shows that Manhattan has the most expensive Airbnb listings across all room types, while the Bronx offers the cheapest options. refer fig(a)

- A significant number of listings are available for 350+ days, followed by those available for 0-20 days. The distribution is bimodal, with peaks at `0 and 365`, while spikes at `90 and 160` days suggest structured hosting availability. refer fig(b)

- Manhattan has the highest mean listing price per bed at `$140`, followed by Brooklyn at `$100` and Queens at `$77`. refer fig(c)

- Hotel listings are the most expensive in Manhattan, while shared and private rooms listings are more affordable. Bronx has the lowest-cost private rooms listings, followed by Queens, while Staten Island offers the cheapest apartment/home listings. refer fig(d)

- The scatter plot shows an inverse relationship between price and reviews—higher-priced listings receive fewer reviews, suggesting they attract a smaller, selective audience. Listings priced between `$1-$750` have the most reviews, reflecting higher booking frequency. refer fig(e)

- Entire homes/apartments have the highest review count at `11,483`, followed by private rooms with `8,766`. Manhattan leads in total reviews due to high footfall, followed by Brooklyn.

- The majority of NYC Airbnb listings are Entire Homes/Apartments and Private Rooms, while Shared Rooms are scarce. Airbnb could boost shared room and hotel listings through incentives, while hosts can capitalize on this market gap. refer fig(g) and fig(h)

- Price and bed count show a strong positive correlation, with prices increasing as bed count rises. Price has a weak negative correlation with minimum nights and a slight positive correlation with availability_365, suggesting higher-priced listings have lower demand but more availability.
